```python
import numpy as np
import random
from datetime import time, timedelta

# Function to generate a random time within a specified range
def generate_random_time(start, end):
    start_seconds = start.hour * 3600 + start.minute * 60 + start.second
    end_seconds = end.hour * 3600 + end.minute * 60 + end.second
    random_seconds = random.randint(start_seconds, end_seconds)
    return str(timedelta(seconds=random_seconds))

# Remove the specified columns
df.drop(columns=['Password', 'Comment', 'Data Limit'], inplace=True)

# Generate "uptime" data based on profile
uptimes = []
for profile in df['Profile']:
    if profile == ' KERJA':
        # Generate time between 10:00:00 and 17:00:00
        uptimes.append(generate_random_time(time(10, 0, 0), time(17, 0, 0)))
    else:
        # Generate time between 20:00:00 and 23:00:00
        uptimes.append(generate_random_time(time(20, 0, 0), time(23, 0, 0)))

# Add the "uptime" column to the DataFrame
df['uptime'] = uptimes

# Display the modified DataFrame
df.head()
```
